```python
import pandas as pd
import matplotlib.pyplot as plt
csv_file='./data.csv'
df = pd.read_csv(csv_file)
```

```python
df.describe()
```

```python
mean_value=df['Calories'].mean()
mean_value2=df['Duration'].mean()
mean_value3=df['Pulse'].mean()
mean_value4=df['Maxpulse'].mean()
```

```python
df['Calories'].fillna(value=mean_value,inplace=True)
df['Duration'].fillna(value=mean_value2,inplace=True)
df['Pulse'].fillna(value=mean_value3,inplace=True)
df['Maxpulse'].fillna(value=mean_value4,inplace=True)
```

```python
import matplotlib.pyplot as plt
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
explode = (0.1, 0, 0, 0,0,0)  
plt.pie(popuratity, explode=explode, labels=languages, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140, wedgeprops = {"edgecolor" : "black",'linewidth' : 1, 'antialiased':True})
plt.axis('equal')
plt.show()
```
